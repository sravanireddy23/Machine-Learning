## Exploratory Data Analysis

**Creating a histogram of the Age.**

```python
plt.figure(figsize=(7,5))
sns.set_style('whitegrid')


plt.hist(ad_data['Age'],bins=30,edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age')

plt.xticks(np.arange(10, 70, 10))
plt.yticks(np.arange(0, 100, 10))

plt.show()
```

**Creating a jointplot showing Area Income versus Age**

```python
g = sns.jointplot(data=ad_data,x='Age',y='Area Income',kind='scatter')
g.ax_joint.set_xticks(np.arange(10, 80, 10))
g.ax_joint.set_yticks(np.arange(10000, 100000, 10000))
plt.show()
```

**Creating a jointplot showing the kde distributions of Daily Time spent on site verses Age**

```python
sns.set_style('whitegrid')
sns.jointplot(data=ad_data,x='Age',y='Daily Time Spent on Site',kind='kde',fill=True)
plt.show()
```
